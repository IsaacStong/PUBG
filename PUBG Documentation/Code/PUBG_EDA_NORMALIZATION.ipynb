```python
def get_colinear(df):
    cor=df.corr()
    col=cor.columns
    row=cor.index
    item1=[]
    item2=[]
    for i in col:
        for j in row:
            if i!=j:
                if abs(cor.loc[i,j]) >= 0.75:
                    item1.append(i)
                    item2.append(j)
    #remove duplicates
    df_col=list(set([tuple(sorted(i)) for i in zip(item1,item2)]))
    return pd.DataFrame({'feature1':[i[0]for i in df_col],'feature2':[i[1]for i in df_col],
                         'Correlation':[cor.loc[i[0],i[1]]for i in df_col]})
```

```python
colinear=get_colinear(train)
```

```python
#get number of players in match
def num_match(df):
    p_in_match=df[['matchId','Id']].groupby('matchId').count().reset_index().rename(columns={'Id':'players_in_match'})
    df=df.merge(p_in_match,on='matchId')
    return df
#get number of players in team
def num_team(df):
    p_in_team=df[['groupId','Id']].groupby('groupId').count().reset_index().rename(columns={'Id':'players_in_team'})
    df=df.merge(p_in_team,on='groupId')
    return df
```

```python
#get total kills in game
def total_game_kills(df):
    tk=df[['matchId','kills']].groupby('matchId').sum().reset_index().rename(columns={'kills':'kills_in_match'})
    df=df.merge(tk,on='matchId')
    return df
#get total kills for team
def total_team_kills(df):
    tk=df[['groupId','kills']].groupby('groupId').sum().reset_index().rename(columns={'kills':'Team_Total_Kills'})
    df=df.merge(tk,on='groupId')
    return df
#get percentage of individual kill in game
def percent_kills(df):
    df['percent_kill']=df['kills']/df['kills_in_match']
    return df
#get total percentage of kills by team
def percent_team_kills(df):
    df['percent_team_kill']=df['Team_Total_Kills']/df['kills_in_match']
    return df
```

```python
#get total damage in game
def total_game_damage(df):
    td=df[['matchId','damageDealt']].groupby('matchId').sum().reset_index().rename(columns={'damageDealt':'damage_in_match'})
    df=df.merge(td,on='matchId')
    return df
#get total damage for team
def total_team_damage(df):
    td=df[['groupId','damageDealt']].groupby('groupId').sum().reset_index().rename(columns={'damageDealt':'Team_Total_damage'})
    df=df.merge(td,on='groupId')
    return df
#get percentage of individual kill in game
def percent_damage(df):
    df['percent_damage']=df['damageDealt']/df['damage_in_match']
    return df
#get total percentage of kills by team
def percent_team_damage(df):
    df['percent_team_damage']=df['Team_Total_damage']/df['damage_in_match']
    return df
#get headshot accuracy/rate
def headshot_rate(df):
    df['headshot_rate']=np.where(df['kills']!=0,df['headshotKills']/df['kills'],0)
    return df
```

```python
#get total items picked up (weapons+heals+boost)
def heal_boosts(df):
    df['heals_and_boosts']=df['boosts']+df['heals']
    return df
def total_items(df):
    df['items']= df['heals_and_boosts']+df['weaponsAcquired']
    return df
```

```python
#get total distance travelled
def total_distance(df):
    df['total_distance']=df['rideDistance']+df['swimDistance']+df['walkDistance']
    return df
```

```python
#Function to do all feature addition/normalization for squads and duos
def do_feature_multi(df):
    df=num_match(df)
    df=num_team(df)
    #kills
    df=total_game_kills(df)
    df=total_team_kills(df)
    df=percent_kills(df)
    df=percent_team_kills(df)
    #damage
    df=total_game_damage(df)
    df=total_team_damage(df)
    df=percent_damage(df)
    df=percent_team_damage(df)
    #headshot rate
    df=headshot_rate(df)
    #items
    df=heal_boosts(df)
    df=total_items(df)
    #distance
    df=total_distance(df)
    return df
```

```python
#Function to do all feature addition/normalization for solo
def do_feature_solo(df):
    df=num_match(df)
    #kills
    df=total_game_kills(df)
    df=percent_kills(df)
    #damage
    df=total_game_damage(df)
    df=total_team_damage(df)
    df=percent_damage(df)
    df=percent_team_damage(df)
    #headshot rate
    df=headshot_rate(df)
    #items
    df=heal_boosts(df)
    df=total_items(df)
    #distance
    df=total_distance(df)
    return df
```

```python
do_feature_multi(train)
```
