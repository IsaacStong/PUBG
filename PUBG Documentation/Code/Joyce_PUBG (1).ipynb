**REMOVE OUTLIERS**

in most cases, outliers will harm the model.


```python
cols = ['assists', 'boosts', 'damageDealt', 'DBNOs', 'headshotKills', 'killPlace', 'killPoints', 'kills', 'killStreaks', 'longestKill', 'matchDuration', 'maxPlace', 'rankPoints', 'rideDistance', 'roadKills', 'swimDistance', 'walkDistance', 'weaponsAcquired', 'winPoints', 'winPlacePerc' ]
```

```python
Q1 = df_solo[cols].quantile(0.05)
Q3 = df_solo[cols].quantile(0.95)
IQR = Q3 - Q1

df_solo = df_solo[~((df_solo[cols] < (Q1 - 1.5 * IQR)) |(df_solo[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df_solo.describe()
```

```python
Q1 = df_duos[cols].quantile(0.05)
Q3 = df_duos[cols].quantile(0.95)
IQR = Q3 - Q1

df_duos = df_duos[~((df_duos[cols] < (Q1 - 1.5 * IQR)) |(df_duos[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df_duos.describe()
```

```python
Q1 = df_squads[cols].quantile(0.05)
Q3 = df_squads[cols].quantile(0.95)
IQR = Q3 - Q1

df_squads = df_squads[~((df_squads[cols] < (Q1 - 1.5 * IQR)) |(df_squads[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
Q1 = df_events[cols].quantile(0.05)
Q3 = df_events[cols].quantile(0.95)
IQR = Q3 - Q1

df_events = df_events[~((df_events[cols] < (Q1 - 1.5 * IQR)) |(df_events[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
assists_viz= [df_solo['assists'], df_duos['assists'],df_squads['assists'], df_events['assists']]
fig, ax = plt.subplots()
ax.boxplot(assists_viz)
plt.xticks([1,2,3,4],['solo', 'duos', 'squads','events'])
plt.show()
```
